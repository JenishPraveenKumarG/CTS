# Joins

```python
# import pandas

import pandas as pd
```

```python
# Creating dataframe

df = pd.DataFrame({
    'roll_no' : [111,114,112,115,113],
    'name' : ['Ashwin','Girish','Jenish','Arun','Naren'],
    'grade' : ['A','B','C','A','C'],
    'marks' : [22,15,10,23,9],
    'city' : ['Chennai', 'Goa', 'Chennai','Goa','Kerala']
})
```

```python
df.head()
```

```python
# another dataframe city mapping

city_state = pd.DataFrame({
    'city' : ['Goa', 'Goa', 'Chennai','Chennai'],
    'state' : ['Goae', 'Goae', 'Tamil Nadu', 'Tamil Nadu']
})
city_state
```

```python
# left join - left table full ah irukum + right table la irundhu left kum idhukum enna enna common ah iruko ellame eduthukum

df.merge(city_state, how = 'left', on = 'city')
```

```python
# roll number df

roll_no = pd.DataFrame({
    'roll_no':[111,113]
})
roll_no
```

```python
# right join

df.merge(roll_no , how = 'right', on = 'roll_no')
```

```python
placement = pd.DataFrame({
    'roll_no' : [111,112, 160],
    'Company' : ['ABC', 'EFG', 'HCL'],
    'Package' : [8,10,7]
})
placement
```

```python
# outer join or full join - means union of all

df.merge(placement , how = 'outer', on = 'roll_no')
```

```python
# inner join - intersection of both 2

df.merge(placement, how = 'inner')
```
