### How to impute missing values

```python
df.isna().sum()
```

```python
# to get the rows thet have null values

df.loc[df.HourlyWindSpeed.isna() == True]
```

```python
# filling the values with the mean

df.loc[df.HourlyWindSpeed.isna() == True , 'HourlyWindSpeed'] = df.HourlyWindSpeed.mean()
```

```python
df.isna().sum()
```

```python
df.loc[df.HourlySkyWeather.isna() == True]
```

```python
df.HourlySkyWeather.unique()
```

```python
# to fill this kinda categorial values we have to use the mostly occured value

df.HourlySkyWeather.mode()
```

```python
df.HourlySkyWeather.value_counts()
```

```python
# filling the values

df.loc[df.HourlySkyWeather.isna() == True , 'HourlySkyWeather'] = 'CLR'
```

```python
df.isna().sum()
```

```python
# simpler methods

df.HourlySkyWeather.fillna('CLR',inplace = True)
```

```python
# creaying a map

mapping = {
    'Low Fat' : 'LF',
    'Regular' : 'R',
    'LF' : 'LF',
    'reg' : 'R',
    'low fat' : 'LF'
}
```

```python
df.Item_Fat_Content = df.Item_Fat_Content.map(mapping)
```

```python
df.Item_Fat_Content.value_counts()
```
