# Grouping data using Group by

```python
a.first()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
```

# Pivot table

```python
df = pd.read_csv('big_mart_sales.csv')

pd.pivot_table(df, index = 'Item_Type', values = 'Item_MRP', aggfunc='mean')
```

```python
pd.pivot_table(df, index = ['Outlet_Size','Item_Type'], values = 'Item_MRP', aggfunc='mean')
```

# Cross Tab
- used to find the frequency of 2 or more factors

```python
pd.crosstab(df['Outlet_Size'], df['Item_Type'])
```

### Creating new feature using aggregation

```python
avg = df.groupby(['Item_Identifier'])['Item_Visibility'].mean().reset_index()
```
