# Matplotlib and Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Matplotlib basics

```python
height = [150,160,165,185]
weight = [70,80,90,100]

plt.plot(height,weight)
```

```python
plt.plot(height,weight)
plt.title('Realtionship between height and weight')
plt.xlabel('Height')
plt.ylabel('weight')
```

```python
calories_burnt = [65,75,95,99]
plt.plot(calories_burnt)
plt.plot(weight)
```

```python
calories_burnt = [65,75,95,99]
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['calories burnt', 'weight'], loc = 'upper left')
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['calories burnt', 'weight'], loc = 'lower right')
```

```python
# xticks

plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['calories burnt', 'weight'], loc = 'lower right')
plt.xticks(ticks = [0,1,2,3], labels = ['p1' , 'p2', 'p3', 'p4'])
```

```python
# changing plot size

plt.figure(figsize=(15,4))
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['calories burnt', 'weight'], loc = 'lower right')
```

```python
# marker

plt.plot(calories_burnt, marker = 'o')
plt.plot(weight , marker = '*')
plt.legend(labels = ['calories burnt', 'weight'], loc = 'lower right')
```

```python
# color changes
'''
y- --> yellow straight line
y-- --> yellow dotted line
r- red
g - green

'''



plt.plot(calories_burnt, marker = 'o')
plt.plot(weight,'y--', marker = '*')
plt.legend(labels = ['calories burnt', 'weight'], loc = 'lower right')
plt.xticks(ticks = [0,1,2,3], labels = ['p1' , 'p2', 'p3', 'p4'])
```

# original data

```python
df = pd.read_csv('big_mart_sales.csv')
```

```python
df = df.dropna(how = 'any')
```

```python
df.head()
```

### Line chart

```python
import numpy as np
price_by_item = df.groupby('Item_Type').Item_MRP.mean()[:10]

x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

plt.figure(figsize=(14,8))
plt.title('Mean price for each item type')
plt.xticks(label = x, ticks = np.arange(len(x)))
plt.plot(x,y,marker = 'o')
```

# Bar chart

```python
import numpy as np
sales_by_outlet_size = df.groupby('Outlet_Size').Item_Outlet_Sales.mean()

sales_by_outlet_size.sort_values(inplace = True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()


plt.title('Sales by outlet size')
plt.xticks(label = x, ticks = np.arange(len(x)))
plt.bar(x,y,color = ['red', 'orange', 'blue'])
```

# Hiostogram

```python
plt.title("Item Mrp distribution")
plt.xlabel('Mrp')
plt.ylabel = ('Frequency')

plt.hist(df['Item_MRP'], bins = 20, color = 'lightblue')
```

```python
# box plot

data = df[['Item_Outlet_Sales']]

red_dimond = dict(markerfacecolor = 'r', marker = 'o')
plt.boxplot(data.values, labels = ['Item Sales'], flierprops=red_dimond)
```

```python
# scatter plot

plt.scatter(df['Item_Weight'][:200], df['Item_Visibility'][:200])
```

# Seaboen

```python
import seaborn as sns
sns.set()
sns.set(style='darkgrid')

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (10,10)
```

```python
# load the dataset

import pandas as pd

df = pd.read_csv('big_mart_sales.csv')
df.head()
```

```python
# Line plot

sns.lineplot(x = 'Item_Weight', y = 'Item_MRP', data = df[:50])
```

```python
# bar plot

sns.barplot(x = 'Item_Weight', y = 'Item_MRP', data = df[:5])
```

```python
# histogram

sns.distplot(df['Item_MRP'])
```

```python
sns.distplot(df['Item_MRP'], hist = False)
```

```python
# Box plot

sns.boxplot(df['Item_Outlet_Sales'], orient='vertical')
```

```python
sns.boxplot(df['Item_Outlet_Sales'], orient='horizontal')
```

```python
# violen plot

sns.violinplot(df['Item_Outlet_Sales'], orient='vertical', color = 'magenta')
```

```python
# scatter plot

sns.scatterplot(x = 'Item_MRP', y = 'Item_Outlet_Sales', data = df[:200])
```

### Advanced plotting

```python
# cat plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'strip', data = df[:250])
```

```python
# swarm plot - deals with overlapping in cat plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'swarm', data = df[:250])
```

```python
# categorical box plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'box', data = df)
```

```python
# boxen plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'boxen', data = df)
```

```python
# violen plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'violin', data = df)
```

```python
# point plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'point', data = df)
```

```python
# bar plot

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales',kind = 'bar', data = df)
```
